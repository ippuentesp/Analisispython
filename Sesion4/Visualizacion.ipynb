```python
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
```

```python
x = [1,2,3,4,5]
y = [2,6,9,4,11]

plt.plot(x,y)
plt.xlabel('Eje x')
plt.ylabel('Eje y')
plt.title('Grafico de barras')

plt.show()
```

```python
datos = np.random.randn(1000)
sns.histplot(datos,color='pink', kde=True)
plt.show()
```

```python
# Set style
sns.set_style("ticks")

# Create histogram with KDE
sns.histplot(datos, color='pink', kde=True, bins=30, alpha=0.7, linewidth=1)

# Add labels and title
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.title('Histogram with KDE')

# Show plot
plt.show()
```

```python
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

# Generar datos aleatorios
datos = np.random.randn(1000)

# Definir una paleta de colores personalizada de rosado claro a lila
paleta = sns.color_palette("green", as_cmap=True)

# Crear el histograma con color degradado
sns.histplot(datos, kde=True, palette=paleta)
plt.show()
```
